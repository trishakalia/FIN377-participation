```python
import requests

# you can change depending on what you searching 
base_url   = 'http://chroniclingamerica.loc.gov/search/pages/results/'
parameters = '?andtext=lynching&format=json'

url = base_url + parameters
print(url)
r= requests.get(base_url + parameters)
```

```python
r.status_code # the error code for a website when it doesnt work 
# 200 means you got a good result 
```

```python
r
r.json() # means shows what is on the webpage 
search_json = r.json()
search_json.keys() # gets a list of the keys ("elements")
search_json['totalItems'] # access element liek in pandas
```

```python
type(search_json['items'])
len(search_json['items']) # 20 articles and each article assigned to an iten 
# access the first element 
search_json['items'][0].keys() # askk what is in the json
# works on each item 
```

```python
# Q2..... 
url = 'https://api.exchangeratesapi.io/latest?base=EUR'
r = requests.get(url)
r.json().keys()
pd.DataFrame(r.json())
```

```python
# Q3: What was the number of Yen per Euro on Jan 2, 2018?
url = 'https://api.exchangeratesapi.io/history?start_at=2018-01-02&end_at=2018-01-02'
r = requests.get(url)
pd.DataFrame(r.json())
```

```python
r.json()['rates']['2018-01-02']['JPY']# now can get for specific date 
```

```python
#Q4:
url = 'https://api.exchangeratesapi.io/history?start_at=2018-01-02&end_at=2019-12-31'
r = requests.get(url)
df = pd.DataFrame(r.json())
# how to pull dictionary inside another variable 
df['EURJPY'] = df['rates'].apply(lambda x: x.get('JPY'))
df
```
